# technology has the highest **sale**

```python
df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

```python
df.groupby(["Category","Sub-Category"])["Sales"].sum().sort_values(ascending=False)
```

## technology has the highest **profit**

```python
cat_prof=df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
cat_prof
```

```python
df.groupby("Country")["Sales"].sum().sort_values(ascending=False)
```

```python
df.groupby(["City","Region"])["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby(["City","Region"])["Profit"].sum().sort_values(ascending=False).head(10)
```

```python
loss=df[df["Profit"]<0]
loss
```

# these are the top 10 loss making cities

```python
loss.groupby(["City","Region"])["Profit"].sum().sort_values().head(10)
```

```python
loss.groupby(["City","Region","Category","Sub-Category"])["Profit"].sum().sort_values().head(10)
```
